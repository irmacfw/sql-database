```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from src.functions import (drop_irrelevant_columns, 
                            remove_all_punctuation,
                            drop_rows_with_missing,
                            select_columns_for_export,
                            plot_avg_price_per_borough,
                            plot_most_frequent_room_type_per_borough,
                            plot_avg_minimum_nights_per_borough,
                            plot_avg_price_by_availability_bucket,
                            plot_topX_expensive_neighborhoods_min_listings
)

# %load_ext autoreload
# %autoreload 2

# Load data
file_path = r'C:\Users\pablo\vscode101\sql-database\AB_NYC_2019.csv'

# df1 = pd.read_csv(file_path, sep=",")
df1 = pd.read_csv(file_path)

# Quick look at columns and missingness
display(df1.info())
# df.dtypes
# df.nunique()
# df.describe()
# display(df.isna().sum())
# display((df.isna().sum()/df.shape[0])*100)
# df.head()
# Display the first few rows
display(df1.head(5))
```

```python
# Drop unnecessary columns
columns_to_drop = ['id', 'host_id', 'host_name', 'latitude', 'longitude', 'number_of_reviews', 'last_review', 'reviews_per_month', 'calculated_host_listings_count']
df1 = drop_irrelevant_columns(df1, columns_to_drop)
```

```python
# Remove all punctuation and all special characters (keeps only letters, numbers, and spaces)
df1 = remove_all_punctuation(df1, 'name')
```

```python
#Remove any row from df1 where the name column has a missing (null/NaN) value
columns_to_drop = ['name']
df1 = drop_rows_with_missing(df1, columns_to_drop)
```

```python
# Cleaniing final Data Frame Data Analysis & Visualization, removing non numeric columns 

columns = ['neighbourhood_group','neighbourhood','price','minimum_nights','availability_365']
df_analysis = select_columns_for_export(df1, columns)
```

```python
df_analysis
```

```python
# Filter out listings with zero price or zero availability for meaningful analysis
df_analysis_filtered = df_analysis[(df_analysis['price'] > 0) & (df_analysis['availability_365'] > 0)]
```

```python
df_analysis_filtered
```

```python
# Preparing final Data Frame only numeric columns for Data Analysis & Visualization

# Group by neighbourhood_group and neighbourhood, aggregate metrics
group_cols = ['neighbourhood_group', 'neighbourhood']
profit_metrics = df_analysis_filtered.groupby(group_cols).agg(
    avg_price=('price', 'mean'),
    std_price=('price', 'std'),
    var_price=('price', 'var'),
    avg_availability=('availability_365', 'mean'),
    std_availability=('availability_365', 'std'),
    var_availability=('availability_365', 'var'),
    count_properties=('price', 'size')  # counts number of rows/listings per group
).reset_index()

# Calculate a simple profitability score
profit_metrics['profitability_score'] = (
    profit_metrics['avg_price'] * profit_metrics['avg_availability'] / (profit_metrics['std_price'] + 1)
)

# Sort by profitability_score
top_profitable = profit_metrics.sort_values('profitability_score', ascending=False).head(10)
```

```python
# Visualization - Top 10 profitable neighbourhoods
plt.figure(figsize=(12, 7))
sns.barplot(
    y='neighbourhood',
    x='profitability_score',
    data=top_profitable,
    hue='neighbourhood_group',
    dodge=False
)
plt.title('Top 10 Most Profitable Neighbourhoods for Airbnb (Score: Avg Price * Availability / Std Price)')
plt.xlabel('Profitability Score')
plt.ylabel('Neighbourhood')
plt.legend(title='Borough')
plt.tight_layout()
```

```python
# Visualize price/availability variation
plt.figure(figsize=(15, 8))
sns.scatterplot(
    data=profit_metrics,
    x='avg_price',
    y='avg_availability',
    size='count_properties',
    hue='std_price',
    palette='viridis',
    legend='brief'
)
plt.title('Neighbourhoods: Avg Price vs. Availability (Size: #Properties, Color: Price Std Dev)')
plt.xlabel('Average Price')
plt.ylabel('Average Availability (days/year)')
plt.tight_layout()
plt.show()
```
