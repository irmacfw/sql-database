```python
df_merged_neigneighbourhood_id = pd.merge(df2, df1, left_on='name', right_on='neighbourhood', how='right')
df_merged_neigneighbourhood_id.head(5)
```

```python
df3.head(5)
```

```python
df_merged_room_type_id = pd.merge(df3, df_merged_neigneighbourhood_id, left_on='type', right_on='room_type', how='right')
display(df_merged_room_type_id.head(5))
```

```python
df_listings = df_merged_room_type_id[['name_y','neighbourhood_ID','room_type_ID','price','minimum_nights','availability_365']]
df_listings.head(5)
```

```python
display(df_listings.info())
```

```python
import re

df_listings.loc[df_listings.index, 'name_y'] = df_listings['name_y'].apply(
    lambda x: re.sub(r'[^A-Za-z0-9 ]+', '', x) if isinstance(x, str) else x
)
```
